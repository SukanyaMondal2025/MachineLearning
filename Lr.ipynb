```python
import pandas as pd
import numpy as np
file=pd.read_csv("/content/archive (1).zip")
```

```python
file.head()
```

```python
file.tail()
```

```python
file.describe()
```

```python
file.columns
```

```python
file.drop(['Country','Year'],axis=1,inplace=True)
```

```python
file.columns
```

```python
file = file.fillna(file.mean(numeric_only=True))
```

```python
file.isnull().sum()
```

```python
file['Status']
```

```python
from sklearn.preprocessing import OrdinalEncoder
encode=OrdinalEncoder()
file['Status']=encode.fit_transform(file[['Status']])
file['Status']
```

```python
file['Status'].isnull().sum()
```

```python
X=  file.drop(columns=["Life expectancy "])
Y = file["Life expectancy "]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
len(y_train)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
linear_pred=model.predict(x_test)
linear_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_test,linear_pred)

```

```python
mse = np.sqrt(mean_squared_error(y_test, linear_pred))
mse
```
